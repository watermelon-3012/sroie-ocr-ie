# TRAIN DATA PREPROCESSING

```python
from os import path, replace, scandir, makedirs

BASE_DIR = r"raw-data\train"

IMG_DIR = path.join(BASE_DIR, "img")
BOX_DIR = path.join(BASE_DIR, "box")
ENT_DIR = path.join(BASE_DIR, "entities")

DATA_DIR = r"data\train"
makedirs(DATA_DIR, exist_ok=True)

def noext(f):
    return path.splitext(f.name)[0]

jpg_files = sorted(
    [f for f in scandir(IMG_DIR) if f.name.endswith(".jpg")],
    key=lambda f: f.name
)

box_files = sorted(
    [f for f in scandir(BOX_DIR) if f.name.endswith(".txt")],
    key=lambda f: f.name
)

entity_files = sorted(
    [f for f in scandir(ENT_DIR) if f.name.endswith(".txt")],
    key=lambda f: f.name
)

if not (len(jpg_files) == len(box_files) == len(entity_files)):
    raise ValueError("Number of files mismatch")

for i in range(len(jpg_files)):
    f_jpg = jpg_files[i]
    f_box = box_files[i]
    f_ent = entity_files[i]

    if noext(f_jpg) != noext(f_box) or noext(f_box) != noext(f_ent):
        raise ValueError(f"Filename mismatch at index {i}")

    print(f"{i:03d}", f_jpg.name)

    replace(f_jpg.path, path.join(DATA_DIR, f"{i:03d}.jpg"))
    replace(f_box.path, path.join(DATA_DIR, f"{i:03d}.txt"))
    replace(f_ent.path, path.join(DATA_DIR, f"{i:03d}.json"))
```

```python
from os import path, scandir, replace, makedirs

IMG_DIR = path.join(DATA_DIR, "img")
BOX_DIR = path.join(DATA_DIR, "box")
ENT_DIR = path.join(DATA_DIR, "entities")

makedirs(IMG_DIR, exist_ok=True)
makedirs(BOX_DIR, exist_ok=True)
makedirs(ENT_DIR, exist_ok=True)

files = sorted(scandir(DATA_DIR), key=lambda f: f.name)

for f in files:
    name = f.name
    src = f.path

    if name.endswith(".jpg"):
        replace(src, path.join(IMG_DIR, name))

    elif name.endswith(".txt"):
        replace(src, path.join(BOX_DIR, name))

    elif name.endswith(".json"):
        replace(src, path.join(ENT_DIR, name))

print("Done splitting back to 3 folders")
```
